```python
df = df.drop('CUST_ID', axis=1)
df.head()
```

```python
(df == 0).sum().sort_values(ascending=False)
```

```python
import numpy as np

cols_to_fix = df.columns

for col in cols_to_fix:
    df[col] = df[col].replace(0, np.nan)
    df[col] = df[col].fillna(df[col].median())
```

```python
df.columns
```

```python
(df==0).sum().sort_values(ascending=False)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
df.fillna(df.median(), inplace=True)
scaler = StandardScaler()
scaled_df = scaler.fit_transform(df)

scaled_df[:5]
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Calculate inertia
inertia = []
for k in range(1, 11):
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(scaled_df)
    inertia.append(km.inertia_)

# Plot elbow curve
plt.plot(range(1, 11), inertia, marker='o')

# Add constant line (y = some value)
constant_y = inertia[-1] * 1.2   # arbitrary example (20% above last inertia)
plt.axhline(y=constant_y)

plt.xlabel("K Values")
plt.ylabel("Inertia")
plt.title("Elbow Method with Constant Line")
plt.show()
```

```python
k_mean = KMeans(n_clusters=7,random_state=42)
```

```python
df["Cluster"] = k_mean.fit_predict(scaled_df)
```

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(x='BALANCE', y='PURCHASES', hue='Cluster', data=df, palette='viridis', s=100)
plt.title('Clusters of Customers based on Balance and Purchases')
plt.xlabel('Balance')
plt.ylabel('Purchases')
plt.legend(title='Cluster')
plt.show()
```
